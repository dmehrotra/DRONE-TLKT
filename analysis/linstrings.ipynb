```python
import geopandas as gpd
import fiona
import os
import pandas as pd
import gpxpy

df = pd.read_csv('../data/all-flights-manifest.csv')
```

```python
fiona.drvsupport.supported_drivers['kml'] = 'rw' # enable KML support which is disabled by default
fiona.drvsupport.supported_drivers['KML'] = 'rw' # enable KML support which is disabled by default

df['date'] = pd.to_datetime(df['date'])
start_date = '2023-07-24'

di = df[df['date']=='2023-08-03']
```

```python
di['fn'] = di['id'].apply(lambda x: f"/Volumes/easystore/Drones/flights/kml/{x}.kml")
```

```python
compiled_flights = []
for _, row in di.iterrows():
    try:
        d = gpd.read_file(row['fn'], driver='KML')


        d['id'] = row['id']
        d['type'] = row['type']
        d['incident_id'] = row['incident_id']
        d['address_map'] = row['address_map']
        d['date'] = row['date']
        d['time'] = row['time_s']

        compiled_flights.append(d)
    except:
        print("nah")


    
```
